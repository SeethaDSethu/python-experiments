```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tempdata = data.drop(['Geography', 'Gender'], axis=1)
Geography = pd.get_dummies(data.Geography).iloc[:,1:]
Gender = pd.get_dummies(data.Gender).iloc[:,1:]
data = pd.concat([tempdata,Geography,Gender], axis=1)  
data

  
```

```python
X = data.drop(['Exited', 'Surname'], axis=1)
Y = data['Exited']

from sklearn.model_selection import train_test_split

X_train, Y_train, X_test, Y_test = train_test_split(X, Y, test_size=0.2, random_state=21)
```

```python
from sklearn.svm import SVC as svc

model = svc(kernel='rbf', degree=8)

model.fit(X,Y)

predicted_labels = model.predict(Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test, predicted_labels))

print(confusion_matrix(Y_test, predicted_labels))

print(accuracy_score(Y_test, predicted_labels))
```
